```python
def regularex(text):
    import re
    # Pattern to remove colons  between the words, extra spaces, single-character words, and two-character words,line break(\n)
    reg = re.compile(r':|\b\w{1,2}\b|\s{2,}|\n')

    # Applying the regex substitution to each element in the cleaned text column
    text = reg.sub(' ', text).strip()
    return text

#removing of numerical values
def remove_num(text):
    reg=re.compile('\d')
    text=reg.sub('',text)
    return text

# needed to remove extra spaces
def extraspace(text):
    import re
    # Pattern to remove extra spaces
    reg = re.compile(r'\s{2,}')

    # Applying the regex substitution to each element in the cleaned text column
    text = reg.sub(' ', text).strip()
    return text
```

```python
def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
import pandas as pd
from collections import Counter


# Step 1: Combine all text in the 'lemmi_cleaned1' column and count word frequencies
all_words = ' '.join(df_copy1['lemmi_cleaned1']).split()
word_freq = Counter(all_words)

# Step 2: Identify words with frequency less than 3
infrequent_words = {word for word, count in word_freq.items() if count < 3}
print("Words with frequency less than 3:", infrequent_words)

# Step 3: Remove infrequent words and create a new column
def remove_infrequent_words(text):
    words = text.split()
    filtered_words = [word for word in words if word not in infrequent_words]
    return ' '.join(filtered_words)

df_copy1['Processed_text'] = df_copy1['lemmi_cleaned1'].apply(remove_infrequent_words)

# Display the updated DataFrame
print(df_copy1)
```

```python
import pandas as pd
from collections import Counter
import numpy as np


# Helper function to create bigrams
def generate_bigrams(text):
    words = text.split()
    bigrams = [f"{w1}_{w2}" for w1, w2 in zip(words, words[1:])]
    return bigrams

# Define bigrams of interest for each category
positive_bigrams = [
    'food_court', 'good_food', 'self_service', 'food_good', 'good_place',
    'budget_friendly', 'must_visit', 'food_affordable', 'good_service',
    'nice_place', 'affordable_price'
]

negative_bigrams = [
    'wait_time', 'bad_service', 'bad_experience', 'waste_money', 'wait_hour',
    'bad_restaurant', 'poor_service', 'taste_bad', 'service_slow'
]

# Function to count the sum of bigrams in a given list
def count_bigrams(text, bigram_list):
    bigrams = generate_bigrams(text)
    bigram_counts = Counter(bigrams)
    # Sum counts of bigrams present in the bigram_list
    return sum(bigram_counts[bigram] for bigram in bigram_list)

# Creating the new columns by summing counts of respective bigrams
df_copy_new['positive_bigrams_count'] = df_copy_new['Processed_text'].apply(lambda x: count_bigrams(x, positive_bigrams))
df_copy_new['negative_bigrams_count'] = df_copy_new['Processed_text'].apply(lambda x: count_bigrams(x, negative_bigrams))

# Display the updated DataFrame
print(df_copy_new)
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Assuming df_copy_new['Processed_text'] contains the tokenized text data

# Initialize CountVectorizer for both unigrams and bigrams
vectorizer = CountVectorizer(ngram_range=(1, 2))

# Fit the vectorizer on the processed text and transform it to BoW matrix
X_text_bow = vectorizer.fit_transform(df_copy_new['Processed_text'])

# Convert sparse matrix to dense if needed
X_text_bow = X_text_bow.toarray()

# Other features
X_other = df_copy_new.drop(columns=['Processed_text', 'rating', 'Processed_text_length_scaled', 'user_review_no', 'comment_likes'])

# Combine BoW matrix with other features
X = np.hstack((X_text_bow, X_other.values))

# Get feature names: BoW feature names + other feature names
bow_feature_names = vectorizer.get_feature_names_out()  # Feature names for BoW
other_feature_names = X_other.columns.tolist()  # Other feature names
all_feature_names = np.concatenate((bow_feature_names, other_feature_names))

# Normalize features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Define target variable
y = df_copy_new['rating']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

### Random Forest Model ###
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Predict with Random Forest
y_pred_rf = rf_model.predict(X_test)

# Get accuracy and evaluation metrics for Random Forest
print("\nRandom Forest Model Accuracy: {:.2f}".format(accuracy_score(y_test, y_pred_rf)))
print("\nRandom Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))

# Get feature importance from RandomForest
feature_importances_rf = rf_model.feature_importances_

# Create a DataFrame for feature importance and names
rf_feature_importance_df = pd.DataFrame({
    'feature': all_feature_names,
    'importance': feature_importances_rf
})

# Sort features by importance (descending order)
top_rf_features = rf_feature_importance_df.sort_values(by='importance', ascending=False).head(30)
print("\nTop 30 Important Features (Random Forest):")
print(top_rf_features)

### Logistic Regression Model ###
lr_model = LogisticRegression(random_state=42, max_iter=1000)
lr_model.fit(X_train, y_train)

# Predict with Logistic Regression
y_pred_lr = lr_model.predict(X_test)

# Get accuracy and evaluation metrics for Logistic Regression
print("\nLogistic Regression Model Accuracy: {:.2f}".format(accuracy_score(y_test, y_pred_lr)))
print("\nLogistic Regression Classification Report:")
print(classification_report(y_test, y_pred_lr))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_lr))

# Get feature importance from Logistic Regression (absolute value of coefficients)
feature_importances_lr = np.abs(lr_model.coef_[0])

# Create a DataFrame for feature importance and names
lr_feature_importance_df = pd.DataFrame({
    'feature': all_feature_names,
    'importance': feature_importances_lr
})

# Sort features by importance (descending order)
top_lr_features = lr_feature_importance_df.sort_values(by='importance', ascending=False).head(30)
print("\nTop 30 Important Features (Logistic Regression):")
print(top_lr_features)

### Support Vector Machine (SVM) Model ###
# Only works for linear kernel SVM
svm_model = SVC(kernel='linear', random_state=42)
svm_model.fit(X_train, y_train)

# Predict with SVM
y_pred_svm = svm_model.predict(X_test)

# Get accuracy and evaluation metrics for SVM
print("\nSVM Model Accuracy: {:.2f}".format(accuracy_score(y_test, y_pred_svm)))
print("\nSVM Classification Report:")
print(classification_report(y_test, y_pred_svm))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_svm))

# Get feature importance from SVM (absolute value of coefficients)
feature_importances_svm = np.abs(svm_model.coef_[0])

# Create a DataFrame for feature importance and names
svm_feature_importance_df = pd.DataFrame({
    'feature': all_feature_names,
    'importance': feature_importances_svm
})

# Sort features by importance (descending order)
top_svm_features = svm_feature_importance_df.sort_values(by='importance', ascending=False).head(30)
print("\nTop 30 Important Features (SVM):")
print(top_svm_features)

### Gaussian Naive Bayes (GNB) Model ###
gnb_model = GaussianNB()
gnb_model.fit(X_train, y_train)

# Predict with Gaussian Naive Bayes
y_pred_gnb = gnb_model.predict(X_test)

# Get accuracy and evaluation metrics for Gaussian Naive Bayes
print("\nGaussian Naive Bayes Model Accuracy: {:.2f}".format(accuracy_score(y_test, y_pred_gnb)))
print("\nGaussian Naive Bayes Classification Report:")
print(classification_report(y_test, y_pred_gnb))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_gnb))

# As GNB doesn't have a straightforward feature importance method, we use the variance of features
gnb_variances = np.var(X_train, axis=0)

# Create a DataFrame for variances (treated as feature importance)
gnb_feature_importance_df = pd.DataFrame({
    'feature': all_feature_names,
    'importance': gnb_variances
})

# Sort features by importance (descending order)
top_gnb_features = gnb_feature_importance_df.sort_values(by='importance', ascending=False).head(30)
print("\nTop 30 Important Features (Gaussian Naive Bayes):")
print(top_gnb_features)
```
