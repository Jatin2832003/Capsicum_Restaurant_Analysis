```python

!pip install selenium

# Import required modules
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Set up WebDriver options
options = webdriver.ChromeOptions()
options.add_experimental_option("detach", True)

# Define the URL
url = "https://www.google.com/maps/place/Roma's+Cafe+Diner/@25.2812364,83.0007112,17z/data=!4m8!3m7!1s0x398e318d79eedfab:0x8b4981f3565c513f!8m2!3d25.2812364!4d83.0032861!9m1!1b1!16s%2Fg%2F11c1s3ql76?entry=ttu"

# Initialize WebDriver
driver = webdriver.Chrome(options=options)
driver.maximize_window()
driver.get(url)
time.sleep(2)

# Scroll to load all reviews
scrollable_div = driver.find_element(By.XPATH, value='//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]')
total_number_of_reviews = driver.find_element(By.XPATH, value='//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]/div[2]/div/div[2]/div[3]').text.split(" ")[0]
total_number_of_reviews = int(total_number_of_reviews.replace(',', '')) if ',' in total_number_of_reviews else int(total_number_of_reviews)

for i in range(0, (round(total_number_of_reviews / 35 - 1))):
    driver.execute_script('arguments[0].scrollTop = arguments[0].scrollHeight', scrollable_div)
    time.sleep(1.5)

# Wait for the page to load (adjust the sleep time as necessary)
time.sleep(5)

# Find and click all "More" buttons 
more_buttons = driver.find_elements(By.XPATH, "//button[contains(@class, 'w8nwRe kyuRq') and contains(@aria-label, 'See more')]")
for button in more_buttons:
    try:
        # Click the button
        driver.execute_script("arguments[0].click();", button)
        # Wait a bit for the action to complete
        time.sleep(1)
    except Exception as e:
        # Pass to continue even if there's an exception
        pass

# Add additional wait to ensure text is expanded
time.sleep(5)

# Find and click all "English translation" buttons
transl_buttons = driver.find_elements(By.XPATH, "//button[contains(@class, 'kyuRq fontTitleSmall WOKzJe') and .//span[text()='See translation (English)']]")


for button in transl_buttons:
    try:
        # Click the button
        driver.execute_script("arguments[0].click();", button)
        # Wait a bit for the action to complete
        time.sleep(1)
    except Exception as e:
        # Pass to continue even if there's an exception
        pass
# Add additional wait to ensure text to translate
time.sleep(5)


# Get the page source
html_content = driver.page_source

# Close the driver
driver.quit()

```

```python
import pandas as pd
import numpy as np

# Function to merge elements into dictionary
def merge_elements_to_dict(content):
    merged_content = []

    for sublist in content:
        merged_dict = {}
        i = 0
        while i < len(sublist):
            if sublist[i] is not None:
                if isinstance(sublist[i], str) and ':' in sublist[i]:
                    key, value = sublist[i].split(': ', 1)
                    merged_dict[key.strip()] = value.strip()
                    i += 1
                elif i + 1 < len(sublist) and sublist[i + 1] is not None:
                    if isinstance(sublist[i], str) and isinstance(sublist[i + 1], str):
                        merged_dict[sublist[i].strip()] = sublist[i + 1].strip()
                    else:
                        merged_dict[sublist[i]] = sublist[i + 1]
                    i += 2
                else:
                    merged_dict[sublist[i]] = None
                    i += 1
            else:
                i += 1
        merged_content.append(merged_dict)
    
    return merged_content

merged_content = merge_elements_to_dict(content)

# Find all unique keys
all_keys = set()
for d in merged_content:
    all_keys.update(d.keys())

# Create an empty DataFrame
df = pd.DataFrame(columns=list(all_keys))  # Convert set to list

# Populate the DataFrame
rows = []
for d in merged_content:
    row = {key: d.get(key, np.nan) for key in all_keys}
    rows.append(row)

# Use concat to create the DataFrame
df2 = pd.concat([df, pd.DataFrame(rows)], ignore_index=True)

# Display the DataFrame
print(df2)
```

```python
from IPython.display import HTML
import base64  
import pandas as pd  

def create_download_link( df, title = "Download CSV file", filename = "data.csv"):  
    csv = df_combined.to_csv(index =False)
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)

create_download_link(df)
```
